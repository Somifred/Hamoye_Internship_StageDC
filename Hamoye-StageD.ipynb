# **Addressing Global Food Shortage: Analyzing Production and Consumption Data for Sustainable Solutions**

# **Introduction**

The global food shortage is a critical issue affecting millions of people worldwide. Despite advancements in agricultural technology and food distribution networks, many regions still face significant challenges in ensuring food security. Understanding the intricate dynamics between food production and consumption is essential to develop effective strategies to combat this issue. This study leverages a comprehensive dataset on food production and consumption/supply to explore potential solutions to the world food shortage problem.

# **Objective**

The primary objective of this project is to analyze the food production and consumption/supply dataset to identify patterns, trends, and correlations that can inform actionable strategies for alleviating global food shortages. By utilizing appropriate data visualizations, we aim to tell a compelling story that highlights key insights and makes each variable in the dataset purposeful without omitting any crucial information.

# **Data Description**

The dataset includes the following key variables:

  ** Country:** The name of the country where the data was collected.
  **  Item:** The type of food product (e.g., Wheat and products).
  **  Year:** The year of data collection.
  **  Value: **The quantitative measure of the food product, either in production or consumption terms.

**Outlier Detection**

```python
# Simple outlier detection (e.g., using IQR)
Q1 = merged_df[['Value', 'Total_Value']].quantile(0.25)
Q3 = merged_df[['Value', 'Total_Value']].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Define a reasonable range for values, considering outliers as values outside 1.5 * IQR
outliers = (merged_df[['Value', 'Total_Value']] < (Q1 - 1.5 * IQR)) | (merged_df[['Value', 'Total_Value']] > (Q3 + 1.5 * IQR))
```

```python
# Print detected outliers
print("Outliers detected:")
print(merged_df[outliers.any(axis=1)])
```

**The Total Production for each Region.**

```python
# Add a column for region based on country
def categorize_region(country):
    if country in ['Algeria', 'Morocco', 'Egypt']:
        return 'Northern Africa'
    elif country in ['Nigeria', 'Kenya']:
        return 'Western Africa'
    elif country == 'Egypt':
        return 'Eastern Africa'
    else:
        return 'Southern Africa'

merged_df['Region'] = merged_df['Country'].apply(categorize_region)

# Calculate total production by region
region_production = merged_df.groupby(['Region'])['Value'].sum().reset_index()

# Plot using Matplotlib
plt.figure(figsize=(10, 6))
sns.barplot(data=region_production, y='Region', x='Value', orient='h')
plt.title('Total Production by Region')
plt.xlabel('Total Production')
plt.ylabel('Region')
plt.show()
```

**Honey Production by Region and Year**

```python
# Filter the data for Honey production
honey_df = merged_df[merged_df['Item'] == 'Honey']

# Aggregate production by region and year
region_year_honey_production = honey_df.groupby(['Region', 'Year'])['Value'].sum().reset_index()

# Pivot the data for heatmap
heatmap_data = region_year_honey_production.pivot('Region', 'Year', 'Value')

# Create a heatmap using Seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', cbar_kws={'label': 'Honey Production (tons)'})
plt.title('Honey Production by Region and Year')
plt.xlabel('Year')
plt.ylabel('Region')
plt.show()
```
